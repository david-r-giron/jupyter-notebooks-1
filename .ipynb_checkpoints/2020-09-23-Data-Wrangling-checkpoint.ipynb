<h3 id="deal_missing_values">Deal with missing data</h3>
<b>How to deal with missing data?</b>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
        c. replace it based on other functions
    </li>
</ol>

<h3>Replacing by the average (mean)</h3>

```python
# Calculate the average of the column of 
# "normalized-losses" (taken as example)

avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN by mean value"

df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Do the same for column "bore", "stroke", "horsepower" and "peak-rpm"

avg_bore=df['bore'].astype('float').mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)
print("avg_bore: ", avg_bore)

avg_stroke=df['stroke'].astype('float').mean(axis=0)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
print("avg_stroke: ", avg_stroke)

avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
print("avg_horsepower: ", avg_horsepower)

avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True)
print("avg_peak_rpm: ", avg_peak_rpm)
```

```python
df[["normalized-losses", "bore", "stroke", "horsepower","peak-rpm"]]
```

<h3>Dropping whole rows missing crucial data</h3>

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

<h2 id="correct_data_format">Correct data format</h2>
<b>We are almost there!</b>
<p>The last step in data cleaning is checking and making sure that all data is in the correct format (int, float, text or other).</p>

In Pandas, we use 
<p><b>.dtype()</b> to check the data type</p>
<p><b>.astype()</b> to change the data type</p>

```python
df.dtypes
```

<p>As we can see above, some columns are not of the correct data type:</p>

1. Numerical variables should have type 'float' or 'int', and 
2. variables with strings such as categories should have type 'object'. 

<p>For example, 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'.</p>

<p>We have to convert data types into a proper format for each column using the "astype()" method.</p>

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

<h2 id="data_standardization">Data Standardization</h2>

<p><b>Standardization</b> is the process of transforming data into a common format which allows the researcher to make the meaningful comparison.
</p>

<b>Example</b>
<p>Transform mpg to L/100km:</p>
<p>In our dataset, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard</p>
<p>We will need to apply <b>data transformation</b> to transform mpg into L/100km?</p>

<p>The formula for unit conversion is<p>
L/100km = 235 / mpg
<p>We can do many mathematical operations directly in Pandas.</p>

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)

df['city-L/100km'] = 235/df["city-mpg"]
```

```python
df['city-L/100km']
```

```python
# transform mpg to L/100km in the column of "highway-mpg", 
# and change the name of column to "highway-L/100km".

df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
```

```python
df.head()
```

<h2 id="data_normalization">Data Normalization</h2>

<b>Why normalization?</b>
<p>Normalization is the process of transforming values of several variables into a similar range.
</p>

<b>Example</b>
<p><b>Target:</b>would like to Normalize those variables so their value ranges from 0 to 1.</p>
<p><b>Approach:</b> replace original value by (original value)/(maximum value)</p>

```python
# replace (original value) by (original value)/(maximum value)

df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()

df[["length","width","height"]].head()
```

<h2 id="binning">Binning</h2>
<b>Why binning?</b>
<p>
    Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
</p>

<b>Example: </b>
<p>In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it has 57 unique values. What if we only care about the price difference between cars with:</p> 

1. high horsepower, 
2. medium horsepower, and 
3. little horsepower

<p>We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins </p>

```python
# Convert data to correct format

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting a histogram of horsepower
# by default bins=10

plt.hist(df["horsepower"])

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("Horsepower Bins")
```

<b>Adjustments to our histogram</b>

1. We would like 3 bins of equal size bandwidth so we use numpy's linspace(start_value, end_value, numbers_generated function.

2. Since we want to include the minimum value of horsepower we want to set start_value=min(df["horsepower"]).

3. Since we want to include the maximum value of horsepower we want to set end_value=max(df["horsepower"]).

4. Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)

print("bins: ", bins)
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
# We apply the function "cut" the determine 
# what each value of "df['horsepower']" belongs to. 

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
```

```python
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_names, df["horsepower-binned"].value_counts())

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

<h3>Bins visualization</h3>
Normally, a histogram is used to visualize the distribution of bins we created above. 

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

<h2 id="indicator">Indicator variable (or dummy variable)</h2>
<b>What is an indicator variable?</b>
<p>
    An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 
</p>

<b>Why we use indicator variables?</b>
<p>
    So we can use categorical variables for regression analysis in the later modules.

```python
# Example: fuel-type
dv_ft = pd.get_dummies(df["fuel-type"])
dv_ft.head()
```

```python
# change column names for clarity 
dv_ft.rename( columns = {'gas':'fuel-type-gas', 'diesel': 'fuel-type-diesel'}, inplace=True )
dv_ft.head()
```

We now have the value 0 to represent "gas" and 1 to represent "diesel" in the column "fuel-type". We will now insert this column back into our original dataset. 

```python
# merge fuel-type dummy variables into our original dataframe
# axis=1 means along the columns

df = pd.concat([df, dv_ft], axis=1)
```

```python
df.head()
```

```python
# drop original column "fuel-type" from "df"

df.drop("fuel-type", axis = 1, inplace=True)

df.head()
```

Create dummy variables for aspiration

```python
df["aspiration"].unique()
```

```python
dv_asp = pd.get_dummies(df["aspiration"])
dv_asp
```

```python
# rename columns for clarity

dv_asp.rename(columns={'std':'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
dv_asp
```

```python
# merge the new dataframe to the original dataframe

df = pd.concat([df, dv_asp], axis=1)
df.head()
```

```python
# drop the column "aspiration"

df.drop('aspiration', axis=1, inplace=True)
```
